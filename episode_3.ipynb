```python
from qiskit import QuantumCircuit
import numpy as np

qc = QuantumCircuit(2)

qc.h(0)
qc.cx(0,1)

qc.draw(output='mpl')
```

```python
from qiskit.quantum_info import Pauli

ZZ = Pauli('ZZ')
```

```python
qr = QuantumCircuit(5)
counter = 0
while counter < 4:
    qr.h(counter)
    for i in range(counter,4):
        qr.cp(2*np.pi/2**(i+1), i+1,counter)
    counter += 1

qr.draw(output = "mpl")
```
